```python
import PyPDF2
import pandas as pd
import numpy as np

rates = pd.DataFrame(columns=["suburb", "oldhouse", "newhouse", "oldunit", "newunit"])
rates.set_index("suburb", inplace=True)

def strip_text(text):
    return text.replace(",", "").replace("$", "").replace("%", "").replace("-", "")

def fill_table(suburb, strings):
    rates.at[suburb, "oldhouse"] = strip_text(strings[0])
    rates.at[suburb, "newhouse"] = strip_text(strings[1])
    if (strip_text(strings[4]).isnumeric()) & (strip_text(strings[5]).isnumeric()):
        rates.at[suburb, "oldunit"] = strip_text(strings[5])
        rates.at[suburb, "newunit"] = strip_text(strings[4])
        strings = strings[6:]
    else:
        print(strings)
        strings = strings[4:]    
    while (strip_text(strings[0]).isnumeric()) & (len(strings) > 1):
        strings = strings[1:]
    if len(strings) > 1:
        analyse_content(strings)

def analyse_content(string_list):
    suburb = string_list[0]
    try:
        fill_table(suburb, string_list[1:])
    except:
        print(suburb, string_list[1:])

with open("./rates.pdf", "rb") as file:
    pdf = PyPDF2.PdfFileReader(file)
    pages = pdf.getNumPages()
    for page_no in range(pages):
        page = pdf.pages[page_no]
        content = page.extractText()
        refined_content = [x.strip() for x in content.split("\n") if x.strip() != ""]
        refined_content = refined_content[len(refined_content) - refined_content[::-1].index("(%)"):]
        print("analyse page", page_no)
        analyse_content(refined_content)

rates
```
