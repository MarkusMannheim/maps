```python
import geopandas as gpd
import pandas as pd
import topojson as tp
```

```python
mesh_data = gpd.read_file(
    "./MB_2021_AUST_SHP_GDA2020.zip",
    bbox=(148.7, -35.7, 149.5, -34.9)
)
mesh_data = mesh_data.iloc[:, [0, 1, 5, 7, 13, 17, 19]]
mesh_data = mesh_data[mesh_data["AREASQKM21"] > 0]
mesh_data.columns = ["code_mb", "category", "code_sa2", "code_sa3", "state", "area", "geometry"]
mesh_data.plot(
    "category",
    figsize=(20, 20),
    legend=True
)
```

```python
water = gpd.GeoDataFrame(columns=["class", "geometry"])
water.at[0, "class"] = "water"
water.at[0, "geometry"] = mesh_data[mesh_data["category"] == "Water"]["geometry"].unary_union
water = tp.Topology(
    water,
    toposimplify=1e-4
).topoquantize(1e4)
water.to_json("water.topojson")
```

```python
parkland = gpd.GeoDataFrame(columns=["class", "geometry"])
parkland.at[0, "class"] = "park"
parkland.at[0, "geometry"] = mesh_data[mesh_data["category"] == "Parkland"]["geometry"].unary_union
parkland = tp.Topology(
    parkland,
    toposimplify=1e-4
).topoquantize(1e4)
parkland.to_json("park.topojson")
```

```python
act = gpd.GeoDataFrame(columns=["class", "geometry"])
act.at[0, "class"] = "border"
act.at[0, "geometry"] = mesh_data[mesh_data["state"] == "8"]["geometry"].unary_union
act = tp.Topology(
    act,
    toposimplify=1e-4
).topoquantize(1e4)
act.to_json("act.topojson")
```

```python
sals = gpd.read_file(
    "SAL_2021_AUST_GDA2020_SHP.zip",
    bbox=(148.7, -35.7, 149.5, -34.9)
)
sals = sals.iloc[:, [0, 1, 2, 6, 10]]
sals = sals[sals["AREASQKM21"] > 0]
sals.columns = ["code_sal", "name", "state", "area", "geometry"]
qbn = [ "The Ridgeway", "Greenleigh", "Karabah", "Queanbeyan", "Queanbeyan West", "Queanbeyan East", "Crestwood", "Jerrabomberra", "Googong" ]
sals = sals[[
    (
        (
            (sals.at[i, "state"] == "8") & 
            ("Remainder" not in sals.at[i, "name"])
        )
        # | ( sals.at[i, "name"] in qbn)
    )
    for i in sals.index
]]
sals = sals[["code_sal", "name", "state", "geometry"]]
sals.columns = ["code", "name", "state", "geometry"]
sals["class"] = "suburb"
sals = sals[["code", "name", "class", "geometry"]]
sals["code"] = sals["code"].astype("int")
# sals = tp.Topology(
#     sals,
#     toposimplify=1e-4
# ).topoquantize(1e4)
# sals.to_json("cbr.topojson")
```

```python
people = ["Residential", "Commercial", "Education"]
mesh_people = mesh_data[mesh_data["category"].apply(lambda x: x in people)].cx[148.9:149.4, -35.6:-35.1]
mesh_people["class"] = "residential"
mesh_people = mesh_people[["class", "geometry"]]
mesh_people = tp.Topology(
    mesh_people,
    toposimplify=1e-4
).topoquantize(1e4)
mesh_people.to_json("mesh_people.topojson")
```

```python
data = pd.read_csv("./rates.csv", index_col="suburb")
data["houses (change)"] = data["newhouse"] / data["oldhouse"] - 1
data["units (change)"] = data["newunit"] / data["oldunit"] - 1
data = data[["newhouse", "newunit", "houses (change)", "units (change)"]]
data.columns = ["houses (average rates)", "units (average rates)",  "houses (change)", "units (change)"]
geodata = sals
# geodata["name"] = geodata["name"].str.replace(" (ACT)", "", regex=False)
values = ["houses (average rates)", "units (average rates)", "houses (change)", "units (change)"]
for suburb in geodata.index:
    if suburb in data.index:
        for value in values:
            geodata.at[suburb, value] = data.at[suburb, value]
geodata = geodata[values + ["geometry"]]
geodata = geodata[pd.notna(geodata[values[0]])]
geodata.to_file("suburb_rates.geojson", driver="GeoJSON")
geodata = geodata[values[::2] + ["geometry"]]
geodata = tp.Topology(
    geodata.reset_index(),
    toposimplify=1e-4
).topoquantize(1e4)
geodata.to_json("rates.topojson")
```
